```python
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')

print(df.head(5))

df.columns
```

```python
# a. subset

species1=df[df['class']=='Iris-setosa']

species2=df[df['class']=='Iris-versicolor']

species3=df[df['class']=='Iris-virginica']


print(species1.head())
print(species2.head())
print(species3.head())
```

```python
# b. merge subsets

merged_data=pd.concat([species1,species2,species3])

print(merged_data.head())
```

```python
# c. sort data on petal length

sorted_data=df.sort_values(by='petal_length',ascending=False)

print(sorted_data.head())
```

```python
# d. transpose data

transposed_data=df.head(5).T

print(transposed_data)
```

```python
# e. melt data

melted_data=pd.melt(df,id_vars=['class'],value_vars=['sepal_length','sepal_width','petal_length','petal_width'])

print(melted_data.head(5))
```

```python
# f. casting data

casted_data=melted_data.pivot_table(index=None,columns='variable',values='value',aggfunc='sum').reset_index()

print(casted_data)
```
